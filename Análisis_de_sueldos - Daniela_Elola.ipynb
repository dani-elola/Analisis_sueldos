2. Printear porcentaje de participación por región, ordenado de mayor a menor. Ej: 

```
- Ciudad Autónoma de Buenos Aires - 59.4%
- Formosa - 0.04%
```

````python
dict_particip_region = {}
total_particip_region = 0

for case in list_cases_unif:
    region = case.get("Region")
    total_particip_region = total_particip_region + 1
    if region not in dict_particip_region:
        dict_particip_region.update({region: 1})
    else:
        dict_particip_region.update({region: dict_particip_region.get(region) + 1})
            
for key, value in dict_particip_region.items():
    dict_particip_region.update({key: round(((dict_particip_region.get(key)*100)/total_particip_region), 2)})
    


sorted_values_region = sorted(dict_particip_region.values(), reverse = True)
sorted_dict_region = {}

for i in sorted_values_region:
    for k in dict_particip_region.keys():
        if dict_particip_region[k] == i:
            sorted_dict_region[k] = dict_particip_region[k]
            break


for key, value in sorted_dict_region.items():
    print(key + " - " + str(value) + "%")
````

```
Junior: de 0 hasta 2 años.

```

8. Formatear las carreras universitarias:
- Nombres capitalizados
- Reemplazar vocales con tilde por vocales sin tilde.
- Reemplazar `ñ` por `n`
- *lic*, *lic.* por Licenciatura
- *tec*, *tec.* por Tecnucatura
* *cs, *cs.* por Ciencias
* *ed, ed.* por Educación
* Transformaciones que se consideren necesarias

Printear porcentaje según carrera

````python
def normalize(s):
    replacements = (
        ("á", "a"),
        ("é", "e"),
        ("í", "i"),
        ("ó", "o"),
        ("ú", "u"),
        ("ñ", "n"),
        ("Arq.", "Arquitectura"),
        ("Arq", "Arquitectura"),
        ("Licenciatura", "Lic."),
        ("Licenciado", "Lic."),
        ("Tecnico", "Tec."),
        ("Tecnicatura", "Tec."),
        ("Ciencias", "Cs."),
        ("Educacion", "Ed."),
        ("Ingenieria", "Ing."),
        ("Arquitectura", "Arq."),
        ("Profesorado", "Prof."),
        ("Profesor", "Prof."),
        ("Recursos humanos", "RRHH"),
        ("Rrhh", "RRHH"),
        ("Relaciones internacionales", "RRII"),
        ("Rrii", "RRII"),
        ("Relaciones publicas", "RRPP")
    )
    
    for a, b in replacements:
        s = s.replace(a, b).replace(a.upper(), b.upper())
    return s    
````

````python
list_career = []
list_career_debugged = []
career = ""


for case in list_cases_unif:
    career = case.get("Carrera")
    if career != "":
        list_career.append(career)

        
for car in list_career:
    car = str.capitalize(car)
    car = normalize(car.strip())
    
    list_career_debugged.append(car)
    

dict_career = {}


for car in list_career_debugged:
    if car not in dict_career:
        dict_career.update({car: 1})
    else:
        dict_career.update({car: dict_career.get(car) + 1})

total_car = len(list_career_debugged)
        
for key, value in dict_career.items():
    porc = (value*100)/total_car
    dict_career.update({key: round(porc,2)})

    
for key, value in dict_career.items():
    print(key + " - " + str(value))
    
````

````python
cotiz_argentina	= 104.79
cotiz_bolivia = 6.6
cotiz_chile = 813.38
cotiz_colombia = 3958
cotiz_costa_rica = 640.44
cotiz_cuba = 23.99
cotiz_ecuador = 25000
cotiz_el_salvador = 8.7545
cotiz_guatemala	= 7.68
cotiz_honduras = 24.56
cotiz_mexico = 20.81
cotiz_nicaragua = 35.45
cotiz_panama = 1
cotiz_paraguay = 7063.69
cotiz_peru = 3.85
cotiz_puerto_rico = 1
cotiz_rep_dominicana = 57.65
cotiz_uruguay = 43.93
cotiz_venezuela	= 248832.4


import re
import locale
import statistics


locale.setlocale(locale.LC_ALL, 'nl_NL')

def dolarizar_sueldo (sueldo, pais): 
    sueldo_dolarizado = -1
    sueldo_bruto = sueldo.replace("$", "").replace("Bs", "").replace("CLP", "").replace("COP", "").replace("USD", "").replace("pesos", "").replace("AR", "").replace("~", "").replace("ars", "").replace("pesos arg", "")
    try:
        sueldo_normalizado = float(locale.atof(sueldo_bruto))
        if pais == "Argentina":
            sueldo_dolarizado = sueldo_normalizado/cotiz_argentina
        if pais == "Bolivia":
            sueldo_dolarizado = sueldo_normalizado/cotiz_bolivia
        if pais == "Chile":
            sueldo_dolarizado = sueldo_normalizado/cotiz_chile
        if pais == "Colombia":
            sueldo_dolarizado = sueldo_normalizado/cotiz_colombia
        if pais == "Costa Rica":
            sueldo_dolarizado = sueldo_normalizado/cotiz_costa_rica
        if pais == "Cuba":
            sueldo_dolarizado = sueldo_normalizado/cotiz_cuba
        if pais == "Ecuador":
            sueldo_dolarizado = sueldo_normalizado/cotiz_ecuador
        if pais == "El Salvador":
            sueldo_dolarizado = sueldo_normalizado/cotiz_el_salvador
        if pais == "Guatemala":
            sueldo_dolarizado = sueldo_normalizado/cotiz_guatemala
        if pais == "Honduras":
            sueldo_dolarizado = sueldo_normalizado/cotiz_honduras
        if pais == "México":
            sueldo_dolarizado = sueldo_normalizado/cotiz_mexico
        if pais == "Nicaragua":
            sueldo_dolarizado = sueldo_normalizado/cotiz_nicaragua
        if pais == "Panamá":
            sueldo_dolarizado = sueldo_normalizado/cotiz_panama
        if pais == "Paraguay":
            sueldo_dolarizado = sueldo_normalizado/cotiz_paraguay
        if pais == "Perú":
            sueldo_dolarizado = sueldo_normalizado/cotiz_peru
        if pais == "Puerto Rico":
            sueldo_dolarizado = sueldo_normalizado/cotiz_puerto_rico
        if pais == "República Dominicana":
            sueldo_dolarizado = sueldo_normalizado/cotiz_rep_dominicana
        if pais == "Uruguay":
            sueldo_dolarizado = sueldo_normalizado/cotiz_uruguay
        if pais == "Venezuela":
            sueldo_dolarizado = sueldo_normalizado/cotiz_venezuela    
        
        
        return sueldo_dolarizado   
    except:
        return sueldo_dolarizado
````
